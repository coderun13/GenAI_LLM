```python
import numpy as np
def sigmoid(x):
    return 1/(1+np.exp(-x))
    
print(sigmoid(0))
print(sigmoid(10)) 
print(sigmoid(-10)) 
```

```python
def tanh(x):
    return np.tanh(x)

print(tanh(0))
print(tanh(10))
print(tanh(-10)) 
```

```python
def relu(x):
    return np.maximum(0,x)

print(relu(0))
print(relu(10))
print(relu(-10))
```

```python
def leaky_relu(x, alpha=0.01):
    return np.where(x>0,x,alpha*x)

print(leaky_relu(0))
print(leaky_relu(10))
print(leaky_relu(-10))
```
