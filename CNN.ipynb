```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

x_images = np.random.rand(100, 64, 64, 3)
y_images = np.random.randint(0, 2, size=(100, 1))

model_cnn = Sequential()

model_cnn.add(Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(64,64,3)))
model_cnn.add(MaxPooling2D(pool_size=(2,2)))
model_cnn.add(MaxPooling2D(pool_size=(2,2)))
model_cnn.add(Flatten())
model_cnn.add(Dense(units=1, activation= 'sigmoid'))

model_cnn.compile(optimizer='adam',
                  loss='binary_crossentropy',
                  metrics=['accuracy'])
model_cnn.fit(x_images, y_images, epochs=10, batch_size=32)
```
